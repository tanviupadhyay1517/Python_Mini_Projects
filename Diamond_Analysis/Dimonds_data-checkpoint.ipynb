```python
# Plot histogram to understand the distribution of diamond prices
plt.figure(figsize=[12,6])
plt.hist(df1["price"], bins=80, color="purple")
plt.xlabel("Price in USD")
plt.ylabel("Frequency")
plt.title("Diamonds Price Histogram")
plt.show()

# This histogram helps us understand how diamond prices are distributed. Most diamonds have lower prices, and very few diamonds fall into high-price ranges.
```

```python
# To visualize how the price changes with the carat
plt.figure(figsize=[12,6])
plt.plot(df1['price'], df1['carat'])
plt.xlabel("Price")
plt.ylabel("Carat")
plt.title("Carat v/s Price")
plt.show()

# This chart helps us see how diamond price changes with carat. Usually, when carat size increases, the diamond price also increases.
```

```python
# To visulize how the price changes with the incrase in depth
plt.figure(figsize = [12,6])
plt.plot(df1['price'], df1['depth'])
plt.xlabel("Price")
plt.ylabel("Depth")
plt.title("Depth v/s Price")
plt.show()

# This chart helps us see how diamond depth changes with price. Most diamonds have similar depth values, even when prices are different.
```

```python
# To visualize the distribution of diamond cut types
plt.figure(figsize=[12,6])
cutCounts.plot(kind="line", color="orange")
plt.xlabel("Cut Type")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Cuts")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# This chart helps us understand which diamond cut type appears most frequently in the dataset. A higher point on the line means more diamonds belong to that cut category.
```

```python
# To visualize the distribution of diamond color types
plt.figure(figsize = [12,6])
colorCounts.plot(kind='line', color='Green')
plt.xlabel("color Type")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Colors")
plt.grid(axis='y')
plt.show()

# This chart helps us understand the availability of diamonds by color. Some color types are more common, while some color types are less available in the dataset.
```

```python
# To visualize the distribution of diamond clarity types
plt.figure(figsize = [12,6])
clarityCounts.plot(kind="line", color="red")
plt.xlabel("Clarity Types")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Clarity")
plt.grid(axis="y")
plt.show()

# Some clarity types are more common, while some clarity types are less available in the dataset.
```

```python
# To visualize the distribution of diamond cut types using a bar chart
plt.figure(figsize = [12,6])
cutCounts.plot(kind="bar", color="skyblue")
plt.xlabel("Cut Type")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Cuts")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--")
plt.show()

# This bar chart helps us compare diamond cut types easily. A taller bar means that cut type appears more frequently in the dataset.
```

```python
# To visualize the distribution of diamond color types using a bar chart
plt.figure(figsize=[12,6])
colorCounts.plot(kind="bar", color="pink")
plt.xlabel("Color Type")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Color")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--")
plt.show()

# The tallest bar represents the color category with the highest number of diamonds.
```

```python
# To visualize the distribution of diamond clarity types using a bar chart

plt.figure(figsize=[12,6])
clarityCounts.plot(kind="bar", color="lightgreen")
plt.xlabel("Clarity Type")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Clarity")
plt.xticks(rotation=0)
plt.grid(axis="y", linestyle="--")
plt.show()

#
```

```python
# Convert the categories to number,There is an easy method to use, the factorize()

df1['color'] = df['color'].factorize()[0]
df1['cut'] = df['cut'].factorize()[0]
df1['clarity'] = df['clarity'].factorize()[0]
df1.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Which diamond should we buy?
# -------------------------------

buy = df.copy()

# Calculate price per carat
buy['price_per_carat'] = buy['price'] / buy['carat']

# Sort diamonds by lowest price per carat
buy = buy.sort_values(by='price_per_carat', ascending=True)

print("Best value for money:")
print(f"{round(buy['price_per_carat'].min(), 2)} $ per carat")

print("\nTop 3 best value diamonds:")
print(
    buy[['carat', 'cut', 'color', 'clarity', 'price', 'price_per_carat']]
    .head(3)
)

# -------------------------------
# Value counts
# -------------------------------

print("\nCut Counts:")
print(df['cut'].value_counts())

print("\nColor Counts:")
print(df['color'].value_counts())

print("\nClarity Counts:")
print(df['clarity'].value_counts())

# -------------------------------
# Countplot - Cut
# -------------------------------

plt.figure(figsize=(8, 5))
sns.countplot(x='cut', data=df, color='green')
plt.xlabel("Cut")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Cut")
plt.xticks(rotation=0)
plt.show()

# -------------------------------
# Countplot - Color
# -------------------------------

plt.figure(figsize=(8, 5))
sns.countplot(x='color', data=df, color='green')
plt.xlabel("Color")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Color")
plt.xticks(rotation=0)
plt.show()

# -------------------------------
# Countplot - Clarity
# -------------------------------

plt.figure(figsize=(8, 5))
sns.countplot(x='clarity', data=df, color='green')
plt.xlabel("Clarity")
plt.ylabel("Frequency")
plt.title("Distribution of Diamond Clarity")
plt.xticks(rotation=0)
plt.show()

# -------------------------------
# Pie Chart - Cut
# -------------------------------

cut_counts = df['cut'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    cut_counts,
    labels=cut_counts.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Diamond Cut Distribution")
plt.show()

# -------------------------------
# Pie Chart - Color
# -------------------------------

color_counts = df['color'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    color_counts,
    labels=color_counts.index,
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Diamond Color Distribution")
plt.show()
```
