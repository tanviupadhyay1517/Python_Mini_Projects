```python
# 27. Check whether a number is an Armstrong number.
num = int(input("Enter a number"))
temp = num
sum = 0

digits = len(str(num))

while temp > 0:
    digit = temp % 10
    sum += digit ** digits
    temp = temp // 10

if num == sum:
    print("Amstrong Number")
else:
    print("Not Amstrong Number")
```

```python
# 33. Take a number and calculate the sum of the factorials of its digits.
num = int(input("Enter a number"))
temp = num
total = 0

while temp > 0:
    digit = temp % 10

    fact = 1
    for i in range(1, digit + 1):
        fact *= i

    total += fact
    temp = temp // 10
print("Sum of factorials:", total)
```

```python
# 34. The computer chooses a random number between 1 and 50. User keeps guessing until correct. Give hints if the guess is higher or lower.
import random

secret = random.randint(1, 50)

while True:
    guess = int(input("Enter your guess (1-50):"))

    if guess == secret:
        print("correct! You guessed it!")
        break
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print("Too low! Try again.")
```

```python
# 35. Take an integer and count how many digits are even and how many are odd.
num = int(input("Enter a number:"))

even = 0
odd = 0

while num > 0:
    digit = num % 10

    if digit % 2 == 0:
        even += 1
    else:
        odd += 1

    num = num // 10

print("Even digits:", even)
print("Odd digits:", odd)
```

```python
# 38. Take a number and sum all digits that are prime (2, 3, 5, 7).
num = int(input("Enter a number:"))
total = 0

while num > 0:
    digit = num % 10

    if digit in (2, 3, 5, 7):
        total += digit
    num = num // 10

print("Sum of prime digits:", total)
```

```python
# 39. Print a diamond-shaped star pattern for a given odd number N.
n = int(input("Enter an odd number:"))

for i in range(1, n+1, 2):
    spaces = (n - i) // 2
    print(" " * spaces + "*" * i)

for i in range (n-2, 0, -2):
    spaces = (n - i) // 2
    print(" " * spaces + "*" * i)
```

```python
# 40. Print a rectangle of size M × N with stars, but replace the diagonal with numbers 1 to min(M,N).
m = int(input("Enter row (M):"))
n = int(input("Enter columns (N):"))

for i in range(m):
    for j in range(n):
        if i == j:
            print(i + 1, end =" ")
        else:
            print("*", end=" ")
    print()
```
